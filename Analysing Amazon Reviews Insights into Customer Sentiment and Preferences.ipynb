```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import LSTM, Embedding, Dense, SpatialDropout1D
from keras.utils import to_categorical
import matplotlib.pyplot as plt

# Step 1: Read the dataset
df = pd.read_csv("D:\\M.sc IT\\MSC-IT-4 YEAR\\SEM-8\\Data Science\\FDS\\sample_dataset_amazonreview.csv")



# Step 2: Split the dataset into features and target variable
X = df['Review_Detailes']
y = df['polarity']

# Step 3: Encode the target variable
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y = to_categorical(y)

# Step 4: Tokenize the text data
max_words = 1000
tokenizer = Tokenizer(num_words=max_words, split=' ')
tokenizer.fit_on_texts(X)
X_tokenized = tokenizer.texts_to_sequences(X)

# Step 5: Pad sequences
max_len = 100  # Adjust as needed
X_padded = pad_sequences(X_tokenized, maxlen=max_len)

# Step 6: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_padded, y, test_size=0.2, random_state=42)

# Step 7: Define the LSTM model architecture
embedding_size = 128
lstm_units = 64

model = Sequential()
model.add(Embedding(max_words, embedding_size, input_length=max_len))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(lstm_units, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(2, activation='softmax'))  # Assuming 2 classes for polarity

# Step 9: Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 10: Train the model
batch_size = 32
epochs = 5
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test))

# Step 11: Evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1] * 100))



history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test))

# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot training and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import LSTM, Embedding, Dense, SpatialDropout1D
from keras.utils import to_categorical
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Step 1: Read the dataset
df = pd.read_csv("D:\\M.sc IT\\MSC-IT-4 YEAR\\SEM-8\\Data Science\\FDS\\sample_dataset_amazonreview.csv")

# Step 2: Preprocess the data

# Remove punctuation
translator = str.maketrans('', '', string.punctuation)
df['Review_Detailes'] = df['Review_Detailes'].apply(lambda x: x.translate(translator))

# Tokenize and remove stopwords
stop_words = set(stopwords.words('english'))
df['Processed_Text'] = df['Review_Detailes'].apply(lambda x: ' '.join([word for word in word_tokenize(x.lower()) if word.isalnum() and word not in stop_words]))

# Step 3: Split the dataset into features and target variable
X = df['Processed_Text']
y = df['polarity']

# Step 4: Encode the target variable
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y = to_categorical(y)

# Step 5: Tokenize the text data
max_words = 1000
tokenizer = Tokenizer(num_words=max_words, split=' ')
tokenizer.fit_on_texts(X)
X_tokenized = tokenizer.texts_to_sequences(X)

# Step 6: Pad sequences
max_len = 100  # Adjust as needed
X_padded = pad_sequences(X_tokenized, maxlen=max_len)

# Step 7: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_padded, y, test_size=0.2, random_state=42)

# Step 8: Define the LSTM model architecture
embedding_size = 128
lstm_units = 64

model = Sequential()
model.add(Embedding(max_words, embedding_size, input_length=max_len))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(lstm_units, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(2, activation='softmax'))  # Assuming 2 classes for polarity

# Step 9: Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 10: Train the model
epochs = 5  # Adjust as needed
batch_size = 32  # Adjust as needed
history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test))

# Step 11: Evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1] * 100))

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
#sentiment analysis

import pandas as pd
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob

# Read the dataset into a pandas DataFrame
df = pd.read_csv("D:\\M.sc IT\\MSC-IT-4 YEAR\\SEM-8\\Data Science\\FDS\\sample_dataset_amazonreview.csv")

# Preprocessing
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    words = word_tokenize(str(text).lower())
    words = [word for word in words if word.isalnum() and word not in stop_words]
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(lemmatized_words)

df['Processed_Text'] = df['Review_Detailes'].apply(preprocess_text)

# Sentiment analysis using TextBlob
def analyze_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

df['sentiment'] = df['Processed_Text'].apply(analyze_sentiment)

# Plot sentiment distribution
plt.figure(figsize=(8, 6))
df['sentiment'].plot(kind='hist', bins=50, color="skyblue", edgecolor="black")
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```
